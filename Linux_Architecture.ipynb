### The Shell 
How you will communicate with the system; cmd line interpreter. 

The shell is a CLI. 

Allows users to give commands to the kernel and receive responses from it. 

### FHS/Filesystem Hierarchy Standard

- How data is stored in a system; organizes data 

- Specifies location where data is stored in the operating system

**Directory:** File that organizes where other files are stored. They are also called "folders". 
- Can contain files and other directories. 

### Kernel 

Component that manages processes and memory. It communciates with the hardware to execute the commands sent by the shell. 

The kernel uses **drivers** to enable applications to execute tasks. 

It also tells the hardware where to save new files, etc.

### Hardware: Physical Components 

**Peripheral Devices:** Hardware components attached and controlled by the computer system . 

Not core devices. 

**Internal Hardware**: Components required to run the computer. Includes main circuit board and all components atttached to it. 

The main circuit baord is called the motherboard. 

- CPU: Main processor; executes instructions provided by programs which enables these programs to run. 
- RAM: Hardware component used for short-term memory; where data is stored temporarily as you perform tasks on your computer. 
- Hard Drive
