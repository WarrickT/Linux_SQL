## Introduction to databases

**Database:** Organized collection of information or data. It's *like* spreadsheets. 

Can be accessed by multiple people simultaneously 

**Relational Database**: Strucutred database contianing tables that are related to each other. 

Each table contains fields of information. 

Relational Databases often have multiple tables. 

**Record/Rows**

**Primary Key** A column where every row has a UNIQUE ENTRY. This can be the id. 

**Foreign Key** A column in table that is a PRIMARY KEY IN ANOTHER TABLE. This CAN be blank and be duplicates. 

**IMPORTANT CONSTRAINT:** A table can only have 1 primary key, but multiple foreign keys. 

## Querying

SQL (Structured Query Language): Used to create, interact with, and request information from a database. 

**Query**: A request for data from a database table or a combination of tables. 

**Log**: A record of events that occur in a organization

## Basic Queries 

**SELECT** Which column to return. For example, you can return the *customerid* column from the *Chinook* database with

SELECT customerid

**FROM** Which table to choose from. SELECT always comes with the *FROM* keyword. 

SELECT employee_id, device_id

FROM employees; 


Note that keywords are not case sensitive 

Semi colons must also be placed at the end of the statement

**SELECT\***: Selects all columns from the specified table. 

**ORDER BY**: Sequences the records returned by a query based on a specified column or columns. This can be *either ascending or descending order*, but by default it's ascending. 

Note that for multiple arguments, it will sort the **first order**, then the **second**, and so on. 

*For example*

SELECT customerid, city, country 

FROM customers 

ORDER BY city; 

If you want to do descending, you will add the **DESC** keyword. 





## Basic Filtering, Clause, and Operators 

This is selecting based on a condition. 

### Basic SQL Operators 

Simply add an extra line with a *WHERE* clause. 

**=** Equal sign. 

**WHERE**: This indicates the condition for a filter. 

For example, *WHERE country = CANADA*

We can also search by pattern. 

**LIKE** Used instead of "=" to apply wildcards to the filter. Used with WHERE to search for a pattern in the column. It's used instead of an equal sign

**% (Wildcards)**: For example 'East%' finds everything that starts with "East"

## SQL Joins

Useful when you information from two different tables 

Let's say you have one about different machines and another about vulnerabiliites of the system 

### INNER JOIN 

table.column

employees.employee_id and machines.employee_id

Use the Primary Key in one table. Employee_id is a Primary Key in employees, but a foreign key in the machines. 

**INNER JOIN** : Returns rows matching on a specified column that exists in more than one table



This includes all columns!

**NULL**: Any non-assigned values. 

employee_id is common, but we don't want to show them. 

**SYNTAX!!!**

SELECT username, office, operating_system 

FROM employees 

INNER JOIN machines ON employees.employee_id = machines.employee_id;


Join machines with the employees table
